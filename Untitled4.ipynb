```python

# 19237 . 어른 상어



def minus(i):
    return int(i) - 1

# # K 는 냄새사라지는 시간 , Arr 은 현재 상어 위치 , Moving 은 우선순위
# N, M ,K = map(int , input().split())
# Arr = [list(map(int , input().split())) for _ in range(N)]
# init_direction = list(map(minus , input().split()))
# moving_Arr = [list(map(minus , input().split())) for _ in range(4*M)]


import numpy as np


# K 는 냄새사라지는 시간 , Arr 은 현재 상어 위치 , Moving 은 우선순위
N, M ,K = map(int , input().split())

Input = list(map(int , input().split()))
Arr = np.array(Input).reshape(N,N).tolist()

init_direction = list(map(minus , input().split()))

Input = list(map(minus , input().split()))
moving_Arr = np.array(Input).reshape(4*M,4).tolist()

print("초기 방향 : ", init_direction)
print("우선순위 : ", moving_Arr)

units = []

# shark 는 어떤 상어인지 , turn 은 몇 턴 남았는지
Smell_shark = [ [ -1 for _ in range(N) ] for _ in range(N) ] 
Smell_turn = [ [ 0 for _ in range(N) ] for _ in range(N) ] 

# 상어위치를 좌표로 확인하기 위함
Occupy_Arr = [ [ [] for _ in range(N) ] for _ in range(N) ] 

# 데이터 초기화
for x in range(N):
    for y in range(N):
        if Arr[x][y] != 0:
            unit = Arr[x][y] - 1
            units.append([unit , x , y , init_direction[unit]])
            Smell_shark[x][y] = unit
            Smell_turn[x][y] = M
            Occupy_Arr[x][y].append(unit)
            
units = sorted(units, key=lambda x: x[0], reverse=False) 

action_dx = [-1,1, 0 , 0]
action_dy = [0 , 0 , -1 , 1]

# 액션 함수
def move_shark(shark):
    
    x = units[shark][1]
    y = units[shark][2]
    direction = units[shark][3]
    
    #print("현재 상어 / 현재 방향 :  " , shark , direction)
    #print("우선 순위 :  " , moving_Arr[shark*4 + direction])
    # 상어는 아무 냄새도 없는 곳을 좋아해
    for i in moving_Arr[shark*4 + direction]:
        
        new_x = x + action_dx[i]
        new_y = y + action_dy[i]
        new_direction = i
        

        if new_x < 0 or new_x >= N or new_y < 0 or new_y >= N:
            continue
        
        # 냄새없는 곳이 잇으면 이동
        if Smell_shark[new_x][new_y] == -1:
            units[shark][1] = new_x
            units[shark][2] = new_y
            units[shark][3] = new_direction
            Occupy_Arr[x][y] = []
            Occupy_Arr[new_x][new_y].append(shark)
             
            return

    # 없으면 자기 냄새로   
    for i in moving_Arr[shark*4 + direction]:
        
        new_x = x + action_dx[i]
        new_y = y + action_dy[i]
        new_direction = i 
        
        if new_x < 0 or new_x >= N or new_y < 0 or new_y >= N:
            continue
            
        # 냄새없는 곳이 잇으면 이동
        if Smell_shark[new_x][new_y] == shark:
            units[shark][1] = new_x
            units[shark][2] = new_y
            units[shark][3] = new_direction
            Occupy_Arr[x][y] = []
            Occupy_Arr[new_x][new_y].append(shark)
            return
            

end_flag = 0
ans = 0

# 메인 시뮬레이션
for turn in range(1000):
    
    #print("상어들 : " , units)
        
    # 1. 상어가 움직여요
    for idx in range(M):
        
        if units[idx][0] == -1:
            continue
        else:
            shark = units[idx][0]
            move_shark(shark)
        
    # 2. 상어끼리 물어뜯어요
    for x in range(N):
        for y in range(N):
            if len(Occupy_Arr[x][y]) > 1:
                
                # 강한 상어..!
                strong_shark = min(Occupy_Arr[x][y])
                strong_shark_idx = Occupy_Arr[x][y].index(strong_shark)
                
                # 약한 상어는 죽어요...
                for idx in range(len(Occupy_Arr[x][y])):
                    if idx == strong_shark_idx:
                        continue
                    else:
                        weak_shark = Occupy_Arr[x][y][idx]
                        units[weak_shark][0] = -1
                        print("상어 죽어욧")
                        Occupy_Arr[x][y].pop(idx)
                

    
                
    # 3. 상어들이 냄새를 뿌려요
    
    # 시간에 따라 냄새 약해져
    for i in range(N):
        for j in range(N):
             
            if Smell_turn[i][j] > 0:
                Smell_turn[i][j] -= 1
                
                if Smell_turn[i][j] == 0:
                    Smell_shark[i][j] = -1
 
    # 이동한데서 냄새 풍겨요
    for idx in range(M):
        
        unit = units[idx][0]
        
        # 죽은 상어는 냄새가 없다
        if unit == -1:
            continue
            
        # 새로운 냄새 발동     
        x = units[idx][1]
        y = units[idx][2]
        direction = units[idx][3]
        Smell_shark[x][y] = unit
        Smell_turn[x][y] = M
        

        
    # 4. 끝낫어요?
    count = 0
    for idx in range(M):
        if units[idx][0] == -1:
            count += 1
    if count == M -1 :
        end_flag = 1
        ans = turn + 1
        break

   
    
# 정답 출력
if end_flag == 0:
    print(-1)
else:
    print(ans)

```
