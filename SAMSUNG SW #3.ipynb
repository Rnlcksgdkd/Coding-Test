```python

## 13458번 . 시험감독

# 문제 자체는 되게 쉬움 , 예외처리에서 삑날 수 있음
# 예외 처리해야되는 경우 잘 잡았다...!  15분 컷
# 항상 이상한경우를 생각해야하는 습관을 들이자
# 이게 음수가되면어쩌지? 요런 느낌? 특히 if문이나 for 문 돌릴떄
# if 문은 범위 , for 문은 인덱싱 조심

import math

# 입력
N = int(input())
Arr = list(map(int , input().split()))
B , C =  (map(int , input().split()))

# 변수 , b 는 총감독수 , c 는 부감독수
Total = 0
b = N
c = 0

for i in range(N):
    need = Arr[i] - B
    if  need <= 0 :
        continue
    else:
        c = c + math.ceil( need/C )

Total = b + c

# 출력
print(Total)
```

```python

## 15638 . 감시

# 입력
N,M = map(int, input().split())
Arr = [ list(map(int , input().split())) for __ in range(N) ]



# 데이터
ret = 0
num_cctv = 0
type_cctv = []
pos_cctv = []
for i in range(N):
    for j in range(M):
        if Arr[i][j] >= 1 and Arr[i][j] < 6 :
            num_cctv = num_cctv + 1
            type_cctv.append(Arr[i][j])
            pos_cctv.append([i,j])

#print("cctv 숫자 : {} ,  cctv 타입 : {} , cctv 위치 : {}".format(num_cctv, type_cctv , pos_cctv))

# CCTV 회전

# 0 : 오른쪽 (+y) , 1 : 왼쪽 (-y) ,  2 : 위쪽 ( -x) ,  3 : 아래쪽 (+x)
def cctv(type_cctv , dir_cctv):
 
    
    # 1번 타입 
    if type_cctv == 0:
        if dir_cctv == 0:
            return [0]
        elif dir_cctv == 1:
            return [3]
        elif dir_cctv == 2:
            return [1]
        elif dir_cctv == 3:
            return [2]
        
    elif type_cctv == 1:
        if dir_cctv == 0 or dir_cctv == 2:
            return [0,1]
        elif dir_cctv == 1 or dir_cctv == 3:
            return [2,3]

    elif type_cctv == 2:
        if dir_cctv == 0:
            return [0,2]
        elif dir_cctv == 1:
            return [0,3]
        elif dir_cctv == 2:
            return [1,3]
        elif dir_cctv == 3:
            return [1,2]
        
    elif type_cctv == 3:
        if dir_cctv == 0:
            return [0,1,2]
        elif dir_cctv == 1:
            return [0,2,3]
        elif dir_cctv == 2:
            return [0,1,3]
        elif dir_cctv == 3:
            return [1,2,3]        
   
    elif type_cctv == 4:
        return [0,1,2,3]
        

# 사각지대 숫자 찾기 , case 에는 각 
def search4th(case):
    
    ret = 0
    
    # 마킹 검사
    marking = [ [ 0 for _ in range(M) ] for __ in range(N) ]
    
    # cctv 에 대해
    for idx in range(len(case)):
        
        # 볼수 있는 범위 검사
        case[idx]
        type_cctv[idx]
        x = pos_cctv[idx][0]  
        y = pos_cctv[idx][1] 
        
        marking[x][y] = 1
        
        # cctv 타입과 회전에 따라 방향 검사
        #print(type_cctv[idx] , case[idx])
        angle = cctv(type_cctv[idx] - 1 , case[idx])

        
        # 방향에 따라 감시마킹
        for i in angle:
            
            x = pos_cctv[idx][0]  
            y = pos_cctv[idx][1] 
        
        
            # 오른쪽 + y
            if i == 0:
                while( y + 1 < M and Arr[x][y+1] != 6):
                    y = y + 1
                    marking[x][y] = 1
            elif i == 1:
                while( y - 1 >= 0 and Arr[x][y-1] != 6):
                    y = y - 1
                    marking[x][y] = 1
            elif i ==2:
                while( x - 1 >= 0 and Arr[x-1][y] != 6):
                    x = x - 1
                    marking[x][y] = 1
            elif i==3:
                while( x + 1 < N and Arr[x+1][y] != 6):
                    x = x+1
                    marking[x][y] = 1
            
    # 사각지대 숫자 리턴
    for i in range(N):
        for j in range(M):
            if marking[i][j] == 0 and Arr[i][j] != 6:
                ret = ret + 1
                
    return ret


CASE = []

# DFS 서치 함수
# idx 는 CCTV 번호 , dir 은 CCTV 회전 정보
def searchDFS(idx ,  case):
    
    new_case = case.copy()
    
    if idx >= num_cctv:
        CASE.append(new_case)
        #print("케이스 검출 검사 : ", new_case)
        return
    
    # 4방향에 대해
    for i in range(4):
        searchDFS(idx + 1 , new_case + [i])


# 메인문 - 마킹 및 서치


MIN = 10000

searchDFS(0 , [])

for idx in range(len(CASE)):
    ans = search4th(CASE[idx])
    
    if ans < MIN:
        MIN = ans

print(MIN)
```

```python


# 14501 . 퇴사


# 입력
N = int(input())
Arr = [ list(map(int , input().split())) for __ in range(N) ]

# 데이터
MAX = 0




CASE = []

# DFS = 탐색 함수
def searchDFS( day , case ):
    
    new_case = case.copy()

    
    
    # 가능한 다음 상담 idx and  상담좀료일
    idx , end = available_day(day)
    
    if len(idx) == 0:
        ans = 0
        for i in range(len(new_case)):
            ans = ans + Arr[new_case[i]][1]
            CASE.append(ans)
        #print(" case : ", case , ans)
        return
    
    for i in range(len(idx)):
        new_case = case.copy()
        new_case.append(idx[i])
        searchDFS( end[i] , new_case) 
    
# 현재 시점부터 할 수 있는 일들을 찾아주는 함수  
def available_day(day):
    
    idx = []
    end = []
    
    for i in range(N):
        if day <= i and i + Arr[i][0] < N + 1:
            idx.append(i)
            end.append(i + Arr[i][0])
    
    #print("현재 날짜 , 앞으로 가능 한 일들 , 종료일 ", day , idx , end )
    return idx , end
    
    
# 메인함수

searchDFS(0 , [] )

# 최대값 찾기
for i in range(len(CASE)):
    if MAX < CASE[i]:
        MAX = CASE[i] 
        
print(MAX)
```
