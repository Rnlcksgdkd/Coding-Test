```python

## Python 기반 탐색 기법에 대한 함수들

# 모든 탐색은 탐색 이전에 환경 (상태 , 행동) 에 대한 정의가 되있어야함
# 임의 탐색이 아닌 Heuristic 탐색은 평가함수를 설정할 것



## 맹목적 탐색 

# (1). DFS

# (2). BFS

# (3). IDS

# (4). UCS



# # Open : 탐색 할 노드 기억
# Open = []
# # Close : 방문 한 노드 기억
# Close = []
# # Sol : 해 기억
# Solution = []
# # Visit : 방문 횟수 : 탐색 수    , 사실 len(Close) 같음

# # Open_2 : IDS 에서 제한된 depth 를 DFS 로 탐색한 후 다시 탐색할 노드들을 저장하기 위한 변수
# Open_IDS = []

## (1). DFS : 깊이 우선 탐색 
# Open 대신 재귀 호출
# 결과 값을 상태로 기억할지 행동으로 기억할지?
def search_DFS(env, state , depth , wp , limit_depth = 10):
    
   
    # 깊이 제한
    if depth >= limit_depth:
        return
    
    # 목표 상태인지 확인
    if env.check_goal(state):
        print("Search Success")
        print("Solution WayPoint : " , wp)
        #print("Close : ", Close)
        print("방문 횟수 :", len(Close))
       
        env.finish_flag = 1
        return 
    
    if env.finish_flag:
        return
    
    # 방문 기록 저장
    Close.append(state)

    #print(state)


    # 자식 노드 (가능한 행동들) 확장 , DFS 의 경우 재귀함수 호출시 Stack 을 이용
    for action_index in range(env.action_num):
        if env.check_possible_action(state, action_index):
            search_DFS( env, env.action(state, action_index), depth + 1 , wp + [action_index], limit_depth)
    
    # 탐색 실패시 루트 함수 종료시 출력 
    if depth == 1 and not env.finish_flag:
        print("Search Failure")


    
## (2). BFS : 너비 우선 탐색 ,  Whilen 문 구조
# depth 제한을 넣으려면 <state, depth> 구조 를 취해야 함  -->  메모리 커짐
# while 문 구조 
def search_BFS(env, state, wp , limit_depth = 10 ):
    
    Open = []
    Close = []
    
    # 루트 노드 
    wp = wp
    Open.append([state, 0])     # 사실 BFS 에서 depth 제한 말고 depth 는 큰의미 없는듯?
    
    # Open 이 바닥날때까지 계속
    while(len(Open) != 0):
        
        state,depth = Open.pop(0)
    
        # 목표 상태인지 확인
        if env.check_goal(state):
            print("BFS Search Success")
            #print(Close)
            print("BFS 탐색 수 : ", len(Close))
            env.finish_flag = 1
            break 
            
        # 깊이 제한 , DFS 랑 다르게 깊이제한 걸림 = 탐색 실패임
        if depth >= limit_depth:
            print("BFS Search Failure")
            break
   
        
        Close.append([state,depth])
    
        # 자식 노드 확장 , BFS 의 경우 리스트(큐) 를 이용
        for action_index in range(env.action_num):
            next_state = env.action(state, action_index)
            if env.check_possible_action(next_state, action_index):
                #print(state)
                #rint(env.action(state, action_index))
                Open.append([next_state, depth + 1])
        #print("Open", Open )
        

# (1) - 2. DFS : While 문 구현 
def search_DFS2(env, state , wp , limit_depth):
    
    Open = []
    Close = []
    
    # 루트 노드 
    wp = wp
    Open.append([state, 0])     # 사실 BFS 에서 depth 제한 말고 depth 는 큰의미 없는듯?
    flag = 0
    
    
    # Open 이 바닥날때까지 계속
    while(len(Open) != 0):
        
        state,depth = Open.pop(-1)
    
        # 목표 상태인지 확인
        if env.check_goal(state):
            print("DFS Search Success")
            #print(Close)
            print("DFS 탐색 수 : ", len(Close))
            env.finish_flag = 1
            return
            
        # 깊이 제한 , DFS 랑 다르게 깊이제한 걸림 = 탐색 실패임
        if depth >= limit_depth:
            
            continue
   
        
        Close.append([state,depth])
    
        # 자식 노드 확장 , BFS 의 경우 리스트(큐) 를 이용
        for action_index in range(env.action_num):
            next_state = env.action(state, action_index)
            if env.check_possible_action(next_state, action_index):
                #print(state)
                #rint(env.action(state, action_index))
                Open.append([next_state, depth + 1])
        #print("Open", Open )
        
        
    print("Search Failure")

# (3). IDS : 반복적 심층 탐색 
def search_IDS(env, state , wp , max_limit_depth , iter_depth):

    Open = []
    Close = []
    Open_IDS =[]
    
    # 루트 노드 
    wp = wp
    Open.append([state, 0])     # 사실 BFS 에서 depth 제한 말고 depth 는 큰의미 없는듯?
    now_limit_depth = 0
    flag = 0
    
    while(now_limit_depth < max_limit_depth ):
         
        # now limit 범위를 다 돌렸으면 depth 확장
        if len(Open) == 0 and len(Open_IDS) != 0 :
            for i in range(len(Open_IDS)):
                Open.append(Open_IDS[i])
            now_limit_depth += iter_depth
        
        # Open 이 바닥날때까지 계속
        while(len(Open) != 0):
            #print("Open", Open )
            state,depth = Open.pop(-1)

            # 목표 상태인지 확인
            if env.check_goal(state):
                print("IDS Search Success")
                #print(Close)
                print("IDS 탐색 수 : ", len(Close))
                flag = 1
                return
            
            if depth + 1 >= now_limit_depth:
                for action_index in range(env.action_num):
                    next_state = env.action(state, action_index)
                    if env.check_possible_action(next_state, action_index):
                        #rint(env.action(state, action_index))
                        Open_IDS.append([next_state, depth + 1])
                continue

            Close.append([state,depth])

            # 자식 노드 확장 , BFS 의 경우 리스트(큐) 를 이용
            for action_index in range(env.action_num):
                next_state = env.action(state, action_index)
                if env.check_possible_action(next_state, action_index):
                    #rint(env.action(state, action_index))
                    Open.append([next_state, depth + 1])
           
    print("Search Failure")
    
## (4). UCS : 균일 비용 탐색
# 균일 비용 탐색 -  Cost 에 대한 정보가 필요함 , 여기서는 Cost = Action num = depth 로 가정 , 
def search_UCS(env, state , wp , max_limit_depth):
    
    Open = []
    Close = []
     
    # 루트 노드 
    wp = wp
    Open.append([state, 0])     # 사실 BFS 에서 depth 제한 말고 depth 는 큰의미 없는듯?
    now_limit_depth = 0
    flag = 0
    
    # Open 이 바닥날때까지 계속
    while(len(Open) != 0):
        state,depth = Open.pop(0)

        # 목표 상태인지 확인
        if env.check_goal(state):
            print("UCS Search Success")
            #print(Close)
            print("UCS 탐색 수 : ", len(Close))
            return
        
        Close.append([state,depth])

        # 자식 노드 확장 , BFS 의 경우 리스트(큐) 를 이용
        for action_index in range(env.action_num):
            next_state = env.action(state, action_index)
            if env.check_possible_action(next_state, action_index):
                #rint(env.action(state, action_index))
                Open.append([next_state, depth + 1])
        
        #print("노드확장 후 Open", Open )
        # 정렬
        Open = sorted(Open, key=lambda Open: Open[1])

        remove = []
        # 중복 값 삭제를 위한 마킹
        for i in range(len(Open)):
            s = Open[i][0]
        
            for j in range(len(Open) - 1, i, -1):
                ns = Open[j][0]
                if s == ns:
                    remove.append(j)
        
        # 중복 값 삭제       
        #!print(remove)
        remove.sort(reverse = True)
        for d in remove:
            Open.pop(d)
        
        #print("정렬 후 Open", Open )
           
    print(" UCS Search Failure ")

```

```python
## 탐색 기법을 적용 Test 해보기 위한 환경들 정의


# 환경은 STATE , ACTION , GOAL , 평가함수(필요하면) 에 대해 정의 

# (1). 2차원 좌표계 x,y 에서 목표 지점 찾기
class XY_GoalPoint():
    
    def __init__(self, goalPoint):
        
        # gp 는 도착지점의 좌표 설정 , finish_flag 는 목표 도착시 다른 탐색을 즉시 종료 
        self.gp = goalPoint
        self.finish_flag = 0
        
        ## State 영역 정의
        self.WIDTH = 10
        self.HEIGHT = 10
        
        
        ## Action 영역 정의
        
        # 0 은 상하좌우 ,  1은 체스-"나이트" 움직임
        self.action_mode = 0
        
        # 상하좌우
        if self.action_mode == 0:
            self.action_x = [0, 1, 0, -1]
            self.action_y = [1, 0, -1, 0]
        # 상하좌우(4) * 대각선 좌우(2) = 8
        elif self.action_mode == 1:
            self.action_x = [2, 2, 1, -1, -2, -2, -1, 1]
            self.action_y = [1, -1, -2, -2, -1, 1, 2, 2]
            
        self.action_num = len(self.action_x)
        
        
    # 행동 체크함수
    def check_goal(self, state):
        
        if state[0] == self.gp[0] and state[1] == self.gp[1]:
            return True
        else:
            return False
        
    def check_same_state(self, state1, state2):
        
        if state1[0]  == state2[0] and state1[1] == state2[1]:
            return True
        else:
            return False
        
    # 행동 체크함수
    def check_possible_action(self, state, action_index):
        
        # State 환경을 벗어난 경우
        if state[0] < 0 or state[0] > self.WIDTH - 2:
            return False
        elif state[1] < 0 or state[1] > self.WIDTH - 2:
            return False
        else:
            return True

            
    # 실제 행동 함수  , 리턴 : 행동 결과 상태
    def action(self, state, action_index):

        return [ state[0] + self.action_x[action_index] , state[1] + self.action_y[action_index] ]
    
    
    # 현재 상태를 출력 함수
    def printState(self, state):
        print("상태 : x,y = {},{}".format(state[0], state[1]))
        return
    
    
    pass

# 하노어의 탑
class Top_Hnor():
    pass

# 그래프 문제
class solve_graph():
    pass

```
