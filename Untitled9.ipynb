```python
import re

def solution(new_id):
    new_id = new_id.lower()
    new_id = re.sub('[+~!@#$%^&*()=+[{]', '', new_id)
    new_id = re.sub('[]}:?,<>]', '', new_id)
    
    index = 0;  start = 0
    print(new_id)
    while(1):
        start = new_id.find("." , start)
        if start == -1 or start == len(new_id) - 1 : break
        index = 1
        while(1):

            if start + index >= len(new_id): break
            if new_id[index + start] == ".":    index += 1
            else:   break
        new_id = new_id[0:start + 1] + new_id[start + index:]
        start = start + 1
 
    print(new_id)
    
    if len(new_id) > 0 :
        if new_id[0] == ".": new_id = new_id[1:]   
    if len(new_id) > 0 :
        if new_id[len(new_id) - 1] == ".": new_id = new_id[:len(new_id) - 1]
    if new_id == "":    new_id = "a"
        

    if len(new_id) >= 16:   
        new_id = new_id[0:15]
        if new_id[len(new_id) - 1] == ".":   new_id = new_id[0:14]
    if len(new_id) <= 2:
        while(len(new_id) != 3):
            new_id = new_id + new_id[len(new_id) - 1]
    return new_id
```
