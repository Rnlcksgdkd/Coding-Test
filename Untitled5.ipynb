```python


## 아기상어

#import time as tm
#start = tm.time()

# 16326 . 아기 상어

N = int(input())
#Arr = [ list(map(int , input().split())) for _ in range(N)]

# 테스트 용 입력
Input = list(map(int , input().split()))
import numpy as np
Arr = np.array(Input).reshape(N,N).tolist()

# 상어 [x,y,size , exp]
shark = [ [0,0] , 2 , 0]

# 물고기들 [ [x,y,size] ...... ]
fishes = []

for x in range(N):
    for y in range(N):
        if Arr[x][y] != 0:
            if Arr[x][y] == 9:
                shark[0][0] = x
                shark[0][1] = y
            else:
                fishes.append([ [x,y] ,Arr[x][y]])
                

#action_dy = [1,-1,0,0]
#action_dx = [0,0,1,-1]

action_dx = [1,-1,0,0]
action_dy = [0,0,1,-1]

# 움직일 수 있는지 체크 , 최단거리 리턴 , 이동 불가시 -1 리턴
def check_move(sp , gp , limit_dist):
    
    pos = [sp[0] , sp[1]]
    distance = 0
    marking_Arr = [ 0 for _ in range(N*N)]
    
    marking_Arr[pos[0]*N + pos[1] ] = 1
    que = [pos] 
    
      
   # BFS
    min_distance = 1000
    
    while(len(que) != 0 ):
        
        node = que.pop(0)
        x = node[0]
        y = node[1]
        
        if distance > min_distance or distance > limit_dist:
            continue

        
        if x == gp[0] and y == gp[1]:
            
            if min_distance > distance:
                min_distance = distance
                #print("Min_distance: ", distance)
    
        else:
            for i in range(4):
                new_x = x + action_dx[i]
                new_y = y + action_dy[i]
            
                # 공간을 넘어가는 경우
                if new_x < 0 or new_x >= N or new_y < 0 or new_y >= N:
                    continue         
                # 이미 방문한 경우
                if marking_Arr[new_x*N + new_y] == 1:
                    continue
                # 물고기한테 가로막힌경우
                if Arr[new_x][new_y] > shark[1]:
                    continue
                
                # 큐에 추가
                marking_Arr[new_x*N + new_y] = 1
                que.append( [new_x , new_y]  )
                
            distance += 1
    
   
    
    if min_distance == 1000:
        #print("탐색 실패 ,BFS 시간 : ", t  )
        return -1
    
    else:
        #print("BFS 탐색 : 최단거리 / 시간  : {} , {}    //  sp , gp : {} . {} ".format(min_distance , t , sp , gp))
        return min_distance

# 먹이를 찾아보자..!
def find_fish():
    
    find_flag = 0
    
    # 조건에 맞는 가장 최단거리의 먹이를 찾자
    target_dist = 100000000
    target_list = []
    target_fish = -1
    
    
    for idx in range(len(fishes)):
        
        fish_pos = fishes[idx][0]
        fish_size = fishes[idx][1]
        
        
        # 못 먹어..!
        if shark[1] <= fish_size:
            continue
        else:
            
            dist = check_move(shark[0] , fish_pos , target_dist)
            #print("후보 타겟 물고기 / 거리: " , fishes[idx] , dist)
            
            if dist != -1:
                find_flag = 1
                
                if dist == target_dist:
                    target_list.append(idx)
                    target_dist = dist
                    
                elif dist < target_dist:
                    target_list = [idx]
                    target_dist = dist
    
    if find_flag == 0:
        return -1
    
    # 당첨
    food_fish = target_list[0]
    #print("당첨 음식 : " , food_fish)
    food_pos = fishes[food_fish][0]
    
    Arr[shark[0][0]][shark[0][1]] = 0
    Arr[food_pos[0]][food_pos[1]] = 9
    shark[0] = [ fishes[food_fish][0][0] , fishes[food_fish][0][1] ] 
    fishes.pop(food_fish)
    
    
    
    return target_dist
    
                    
                    


# 메인 시뮬레이션

end_flag = 0
time = 0

while(not end_flag):
    
    #print("상어와 물고기 : " , shark , fishes)
    #print("공간 : ", Arr)
    dist = find_fish()
    
    print(fishes)
    
    if dist == -1:
        break
    else:
        

        # 경험치 업 / 진화
        shark[2] += 1
        
        # 진화
        if shark[2] >= shark[1]:
            shark[2] = shark[2] - shark[1]
            shark[1] += 1
        

        #print(dist)
        time += dist
        
    
print(time)

#print()
#print("전체 실행 시간 : " , tm.time()-start)

```
