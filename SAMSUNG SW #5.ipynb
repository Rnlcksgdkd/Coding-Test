```python

# 14503. 로봇 청소기

# 입력 처리
N , M = map(int , input().split())
x,y,dir = map(int , input().split())
Arr = [list(map(int,input().split())) for __ in range(N)]

# 마킹

ret = 0

action_dx = [-1,0,1,0]
action_dy = [0,1,0,-1]

# 왼쪽 방향으로 회전
def rotate_left(dir):
    if dir == 0:
        return 3
    elif dir == 1:
        return 0
    elif dir == 2:
        return 1
    elif dir == 3:
        return 2

# 후진
def back(dir):
    if dir == 0:
        return 2
    elif dir == 1:
        return 3
    elif dir == 2:
        return 0
    elif dir == 3:
        return 1

    
# 청소기함수 , 청소한 영역은 2로 표시한다
def clean(x,y,dir):
    
    # 1
    Arr[x][y] = 2
    
    next_dir = rotate_left(dir)
    next_x = x + action_dx[next_dir]
    next_y = y + action_dy[next_dir]
    
    # 2 - a
    if next_x >= 0 and next_x < N and \
        next_y >= 0 and next_y < M and Arr[next_x][next_y] == 0:
        clean(next_x,next_y,next_dir)
        return
    
    # 2 - c , 2 - d
    elif Arr[x + 1][y] != 0 and  Arr[x - 1][y] != 0 and \
             Arr[x][y + 1] != 0 and  Arr[x][y - 1] != 0:
        
        # 후진 
        next_dir = back(dir)
        next_x = x + action_dx[next_dir]
        next_y = y + action_dy[next_dir]
        
        # 2 - d
        if Arr[next_x][next_y] == 1:
            return
        else:
            clean(next_x,next_y,dir)
            return
              
    # 2 - b
    elif next_x >= 0 and next_x < N and \
        next_y >= 0 and next_y < M and Arr[next_x][next_y] != 0:
        clean(x,y,next_dir)
        return
        



# 메인단

clean(x,y,dir)

for x in range(N):
    for y in range(M):
        if Arr[x][y] == 2:
            ret = ret + 1

print(ret)


```
