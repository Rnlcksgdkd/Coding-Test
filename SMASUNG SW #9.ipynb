```python

# 17822 . 원판 돌리기

N ,M ,T = map(int , input().split())
arr = [list(map(int, input().split())) for _ in range(N)]
rot = [list(map(int , input().split())) for _ in range(T)]

action_dx = [0 ,0 ,1 ,-1]
action_dy = [1 ,-1 ,0 ,0]

# 인접수를 리스트로 리턴
def nearby(pos , marking):

    x = pos[0];  y = pos[1]
    val = arr[x][y]
    stack = [pos]
    lst = [pos]

    while(len(stack) != 0 ):
        node = stack.pop(-1)
        x = node[0]
        y = node[1]
        marking[x][y] = 1
        for i in range(4):
            new_x = x + action_dx[i]
            new_y = y + action_dy[i]
            if new_x < 0 or new_x > N - 1:
                continue
            if 0 <= new_y < M and marking[new_x][new_y] == 0 and arr[new_x][new_y] == val:
                lst.append([new_x, new_y])
                stack.append([new_x, new_y])
            elif new_y == -1 and marking[new_x][M-1] == 0 and arr[new_x][M - 1] == val:
                lst.append([new_x, M - 1])
                stack.append([new_x, M - 1])
            elif new_y == M and marking[new_x][0] == 0 and arr[new_x][0] == val:
                lst.append([new_x, 0])
                stack.append([new_x, 0])
    return lst

# 인접수를 다 지워버려 (지우면 -1을 넣어주자)
def del_nearby():
    flag = 1
    marking_arr = [[0]*M for _ in range(N)]

    for x in range(N):
        for y in range(M):
            if arr[x][y] == -1 or marking_arr[x][y] == 1:
                continue
            lst = nearby([x, y] , marking_arr)
            # 인접하고 같은 숫자 제거
            if len(lst) > 1:
                for pos in lst:
                    arr[pos[0]][pos[1]] = -1
                    if flag == 1:
                        flag = 0
    if flag == 1:
        return -1
    else:
        return 1

def Action(rotation):
    x = rotation[0]; d = rotation[1]; k = rotation[2]

    # 누가 회전해야 되는가
    lst = []
    for i in range(N):
        if (i+1) % x == 0:
            lst.append(i)
    # 원반들 회전
    for idx in lst:
        # 시계
        if d == 0:
            temp = arr[idx][M-k:]
            arr[idx] = arr[idx][0:M-k]
            temp += arr[idx]
            arr[idx] = temp
        # 반시계
        elif d == 1:
            temp = arr[idx][0:k]
            arr[idx] = arr[idx][k:]
            arr[idx] += temp
    # 인접수 찾기
    flag = del_nearby()

    # 없으면 평균 기준으로 가감
    if flag == -1:
        ans = 0
        idx = 0
        for i in range(N):
            for j in range(M):
                if arr[i][j] != -1:
                    ans += arr[i][j]
                    idx += 1
        if idx != 0:
            avg = ans / idx
            for i in range(N):
                for j in range(M):
                    if arr[i][j] != -1:
                        if arr[i][j] > avg:
                            arr[i][j] -= 1
                        elif arr[i][j] < avg:
                            arr[i][j] += 1
# 메인함수
def simulation():
    for i in range(T):
        Action(rot[i])
    ans = 0
    for i in range(N):
        for j in range(M):
            if arr[i][j] != -1:
                ans += arr[i][j]
    return ans

# 결과출력
print(simulation())

```

```python

# 19235 . 모노미노도미노

N = int(input())
arr = [list(map(int, input().split())) for _ in range(N)]

# 초록/파랑 영역
green = [ [0 ] *4 for _ in range(6)]
blue = [[ 0 ] *4 for _ in range(6)]

# 블록 형태 정의
block_x = [[0] , [0 ,0] , [0 ,1]]
block_y = [[0] , [0 ,1] , [0 ,0]]


global score
score = 0

# 특수존 검사
def check_special(arr):
    flag = 0
    flag1 = 0
    for x in range(2):
        for y in range(4):
            if arr[x][y] == 0:
                continue
            if x == 0:
                if arr[x][y] != 0:
                        flag = 1
            else:
                if arr[x][y] != 0:
                        flag1 = 1
    clear_line = flag + flag1
    if clear_line == 0:
        return 0
    elif clear_line == 1:
        for y in range(4):
            arr[5][y] = 0
        for x in reversed(range(5)):
            for y in range(4):
                arr[x + 1][y] = arr[x][y]
                arr[x][y] = 0
        return 1
    elif clear_line == 2:
        for y in range(4):
            arr[5][y] = 0
            arr[4][y] = 0
        for x in reversed(range(4)):
            for y in range(4):
                arr[x + 2][y] = arr[x][y]
                arr[x][y] = 0
        return 2

# 라인클리어 확인 및 클리어 함수
def check_line(arr):

    global score
    clear_flag = 0
    for x in range(6):
        flag = 1
        for y in range(4):
            if arr[x][y] != 0:
                continue
            else:
                flag = 0
                break
        # 블록 지우자
        if flag == 1:
            for y in range(4):
                arr[x][y] = 0
            if x != 0:
                #print(arr)
                for i in reversed(range(x)):
                    for j in range(4):
                        if arr[i][j] == 0:
                            continue
                        add = 1

                        if arr[i][j] == 2:
                            pass
                            while (i + add < 5):
                                if i + add + 1 < 6 and arr[i + add + 1][j] != 0:
                                    break
                                if arr[i + add + 1][j + 1] != 0:
                                    break
                                add += 1
                            arr[i + add][j] = arr[i][j]
                            arr[i][j] = 0
                            arr[i + add][j + 1] = arr[i][j + 1]
                            arr[i][j + 1] = 0

                        else:
                            while(i + add < 5):
                                if i + add + 1 < 6 and arr[i + add + 1][j] != 0:
                                    break
                                add += 1
                            arr[i + add][j] = arr[i][j]
                            arr[i][j] = 0
            score += 1
            clear_flag = 1
    return clear_flag

# 초록이
def move_green(block):
    # 블록 생성
    block_type = block[0] - 1
    x = 0;  y = block[2];
    # 블록이 이동해영
    if block_type == 0:
        while(x < 5):
            if x + 1 < 6 and green[x + 1][y] != 0:
                break
            x = x + 1
        green[x][y] = 1
    elif block_type == 1:
        while(x < 5):
            if x + 1 < 6 and green[x + 1][y] != 0:
                break
            if green[x + 1][y + 1] != 0:
                break
            x = x + 1
        green[x][y] = 2
        green[x][y + 1] = 2
    elif block_type == 2:
        while(x + 1 < 5):
            if x + 2 < 6 and green[x + 2][y] != 0:
                break
            x = x + 1
        green[x][y] = 1
        green[x + 1][y] = 1
    #print(green)
    # 라인 체크

    check_line(green)
    # 특수 존 체크
    check_special(green)

    #print(green)

# 파랑이
def move_blue(block):
    # 블록 생성
    block_type = block[0] - 1
    x = 0;  y = block[1]

    if block_type == 1:
        block_type = 2
    elif block_type == 2:
        block_type = 1

    # 블록이 이동해영
    if block_type == 0:
        while(x < 5):
            if x + 1 < 6 and blue[x + 1][y] != 0:
                break
            x = x + 1
        blue[x][y] = 1
    elif block_type == 1:
        while(x < 5):
            if x + 1 < 6 and blue[x + 1][y] != 0:
                break
            if blue[x + 1][y + 1] != 0:
                break
            x = x + 1
        blue[x][y] = 2
        blue[x][y + 1] = 2
    elif block_type == 2:
        while(x + 1 < 5):
            if x + 2 < 6 and blue[x + 2][y] != 0:
                break
            x = x + 1
        blue[x][y] = 1
        blue[x + 1][y] = 1
    #print(blue)
    # 라인 체크

    check_line(blue)
    # 특수 존 체크
    check_special(blue)

    #print(blue)

# 메인 함수
def simulation():

    global score
    # N턴 진행
    for i in range(N):
        # 게임 진행
        #print("초록 turn " , i + 1)
        move_green(arr[i])
        #print("파랑 turn ", i + 1)
        move_blue(arr[i])
    num = 0
    for x in range(6):
        for y in range(4):
            if blue[x][y] != 0:
                num += 1
            if green[x][y] != 0:
                num += 1
    print(score)
    print(num)

# 결과 출력
simulation()
```
