```python

# 16326 . 아기 상어

N = int(input())
Arr = [list(map(int , input().split())) for _ in range(N)]

# 물고기들 = [ [x,y,size] , [] ..... ]
fishes = []

# 상어 = [x,y,size,exp]
shark = [0,0 , 2 , 0]

# 물고기들 위치 및 상어 정보 초기화
for x in range(N):
    for y in range(N):
        if Arr[x][y] == 9:
            shark[0] = x;  shark[1] = y
        elif Arr[x][y] != 0:
            fishes.append([x,y,Arr[x][y]])

#력시뮬레이션 함수
def simulation():
    
    turn = 0
    
    while(1):
        
        # BFS 좌표마다 최단거리 계산
        dist_Arr = BFS( [shark[0] , shark[1] ] )
        
        min = 999999
        fish_idx = -1
           
        # 어떤 물고기가 먹기 좋을까요?
        idx = -1
        for fish in fishes:
            idx = idx + 1
            x = fish[0] ; y = fish[1]; size = fish[2]
            
            # 큰놈은 못먹어
            if size >= shark[2]:
                continue
                
            # 어떤놈이 가장 가까울까
            if dist_Arr[x][y] != -1 and min > dist_Arr[x][y]:
                min = dist_Arr[x][y]
                fish_idx = idx
        
        # 잡아먹을 물고기가 없으면 끝
        if fish_idx == -1:
            return turn

        # 잡아먹을 물고기가 있다면 진핼
        else:
            
            # 물고기 업데이트 , 물고기 죽어욧
            eat = fishes[fish_idx]
            x = eat[0];  y = eat[1];  size = eat[2]
            
            # Arr 업데이트
            Arr[x][y] = 9
            Arr[shark[0]][shark[1]] = 0        
            
            # fishes 업데이트
            fishes.pop(fish_idx)
            
            # shark 업데이트
            shark[0] = x; shark[1] = y; shark[3] += 1;
            
            # 상어 성장
            if shark[3] >= shark[2]:
                shark[3] -= shark[2]
                shark[2] += 1
                
            # 턴 진행
            turn += min
            
action_dx = [0,0 ,1,-1]
action_dy = [1,-1,0,0]

# BFS 를 통해 각 좌표까지의 최단거리를 계산한다
def BFS(start_point):

    # N*N 배열
    dist_Arr = [ [-1]*N for _ in range(N)]
    que = [start_point]
    dist_Arr[start_point[0]][start_point[1]] = 0
    shark_size = shark[2]
    
    # 탐색
    while(len(que) != 0 ):
        
        node = que.pop(0)
        x = node[0];  y = node[1]
        
        # 확장
        for i in range(4):
            n_x = x + action_dx[i];  n_y = y + action_dy[i]
            
            # 조건 안맞으면 패스
            if -1 < n_x < N and -1 < n_y < N and dist_Arr[n_x][n_y] == -1 and shark_size >= Arr[n_x][n_y]:
                dist_Arr[n_x][n_y] = dist_Arr[x][y] + 1
                que.append([n_x , n_y])
    return dist_Arr

# 결과 출력
ans = simulation()
print(ans)
```

```python
####

# 3190 . 뱀

def minus(i):
    return int(i) - 1

N = int(input())
K = int(input())

apple = [list(map(minus, input().split())) for _ in range(K)]
L = int(input())
turning = [list(map(str, input().split())) for _ in range(L)]
for i in range(len(turning)):
    turning[i][0] = int(turning[i][0])

head = [0, 0, 1]
unit = [[0, 0]]
# 빈공간  0  /  뱀 1   /  사과 2
Arr = [[0] * N for _ in range(N)]
Arr[0][0] = 1
for i in range(len(apple)):
    Arr[apple[i][0]][apple[i][1]] = 2

# 0 위쪽 / 1 오른쪽 / 2 아래쪽 / 4 왼쪽
action_dx = [-1, 0, 1, 0]
action_dy = [0, 1, 0, -1]

# 뱀의 행동 함수
def action():

    x = head[0];  y = head[1];  direction = head[2];

    # 이동
    new_x = x + action_dx[direction];  new_y = y + action_dy[direction]

    # 종료 조건
    if new_x < 0 or new_x > N - 1 or new_y < 0 or new_y > N - 1:
        return -1
    if Arr[new_x][new_y] == 1:
        return -1

    # 사과를 먹었으면
    flag = 0
    if Arr[new_x][new_y] == 2:
            flag = 1

    # 뱀 이동 (머리 확장)
    head[0] = new_x;   head[1] = new_y;
    unit.append([new_x , new_y])
    Arr[new_x][new_y] = 1

    # 사과를 못 먹었으면
    if flag == 0:
        node = unit.pop(0)
        Arr[node[0]][node[1]] = 0





def simulation():
    time = 0
    while (1):
        time = time + 1
        #print(unit)
        #print(Arr)
        flag = action()
        if flag == -1:
            return time

        # 방향 변환 체크
        direction = head[2]

        for i in range(len(turning)):
            if time == turning[i][0]:
                change = turning[i][1]

                # 오른쪽 회전
                if change == "D":
                    if direction + 1 > 3:
                        direction = 0
                    else:
                        direction += 1
                # 왼쪽 회전
                else:
                    if direction - 1 < 0:
                        direction = 3
                    else:
                        direction -= 1

        head[2] = direction


# 결과출력
print(simulation())
```

```python

# 14889 . 스타트와 링크

N = int(input())
Arr = [list(map(int, input().split())) for _ in range(N)]

MIN = 9999999

# 팀의 전체 능력치를 계산합니당
def cal_val(lst):

    ans = 0
    for i in lst:
        for j in lst:
            ans += Arr[i][j]
    return ans


# DFS 완전 탐색
def DFS(idx , team_1 , team_2):

    global MIN
    if idx == N:
        if len(team_1) == len(team_2):

            ans = abs(cal_val(team_1) - cal_val(team_2))

            if ans < MIN:
                MIN = ans
        return

    new_team_1 = team_1.copy()
    new_team_2 = team_2.copy()
    DFS(idx + 1 , new_team_1 + [idx] , new_team_2 )
    DFS(idx + 1 , new_team_1 , new_team_2  + [idx])



# 메인 함수
def simulation():
    DFS(0, [] , [])

# 결과 출력
simulation()
print(MIN)
```

```python
# 16235 . 나무 재테크

N ,M ,K = map(int , input().split())
Arr = [ [5 ] *N for _ in range(N)]
arr = [list(map(int , input().split())) for _ in range(N)]
tree = [list(map(int , input().split())) for _ in range(M)]

# 2차원 배열
Occupy = [ [ [] for _ in range(N)] for _ in range(N)]
for i in range(len(tree)):
    x = tree[i][0] - 1;  y = tree[i][1] - 1
    Occupy[x][y].append(tree[i][2])

# 확장
action_dx = [0 ,0 ,1 ,-1 ,1 ,1 ,-1 ,-1]
action_dy = [1 ,-1 ,0 ,0 ,1 ,-1 ,1 ,-1]

# 나무 함수
def Action():

    # 봄 ~ 여름
    for x in range(N):
        for y in range(N):
            if len(Occupy[x][y]) == 0:
                continue
            elif len(Occupy[x][y]) > 1:
                Occupy[x][y].sort()
            death_tree = []
            death_idx = 0
            # 나무가 양분을 먹고 못먹은 나무는 사망
            for i in range(len(Occupy[x][y])):
                if Arr[x][y] >= Occupy[x][y][i]:
                    Arr[x][y] -= Occupy[x][y][i]
                    Occupy[x][y][i] += 1
                else:
                    death_idx = i
                    death_tree = Occupy[x][y][i:]
                    Occupy[x][y] = Occupy[x][y][0:i]
                    break

            for i in range(len(death_tree)):
                Arr[x][y] += int(death_tree[i]/2)

            # for i in reversed(range(len(death))):
            #     death_age = Occupy[x][y].pop(death[i])
            #     Arr[x][y] += death_age

    # 가을 ~ 겨울 같이 진행
    for x in range(N):
        for y in range(N):
            if len(Occupy[x][y]) > 0:
                for i in range(len(Occupy[x][y])):
                    if Occupy[x][y][i] % 5 == 0:
                        for j in range(8):
                            new_x = x + action_dx[j]
                            new_y = y + action_dy[j]
                            if 0 <= new_x < N and 0 <= new_y < N:
                                Occupy[new_x][new_y].append(1)
            Arr[x][y] += arr[x][y]

# 메인 함수
def simulation():

    ans = 0
    for _ in range(K):
        Action()
    for x in range(N):
        for y in range(N):
            if len(Occupy[x][y]) != 0:
                ans += len(Occupy[x][y])
    return ans

# 결과 출력
print(simulation())
```
