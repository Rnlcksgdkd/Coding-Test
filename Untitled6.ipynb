```python

# 2. 인형들

# PyPy3 는 통과 , Python3 는 못통과.. 파이썬3 쓰면 안되나...?
# sum / len 함수 계속 다시적용해주는거 시간 차이 난다

N , K = map(int , input().split())
lst = list(map(int , input().split()))

MIN = float("inf")
end_flag = 0

if K == 1:
    MIN = 0
    end_flag = 1
    
for i in range(N - K + 1):
    if end_flag == 1:
        break
    sample = [lst[i]]
    
    
    for j in range(1,N - i,1):
        sample.append(lst[i + j])
        
        if j >= K - 1:
            ans = 0
            
            # 표준 편차 계산부분
            SUM = sum(sample)
            LEN = len(sample)
            avg = SUM/LEN
            for k in range(LEN):
                ans += pow((avg -sample[k]),2)
            ans = ans/LEN
            ans = pow(ans , 0.5)

            if ans < MIN:
                MIN = ans
print(MIN)
        



```
