```python

import copy

# 19238 . 스마트 택시

def minus(i):
    ans = int(i) - 1
    return ans



N , M , F = map(int , input().split())
Arr = [list(map(int , input().split())) for _ in range(N)]
now_pos = list(map(minus , input().split()))

# 1이상 N 이하 (원소 4개짜리 M 명)
clients = [list(map(minus , input().split())) for _ in range(M)]

action_dx = [1, -1 , 0 ,0]
action_dy = [0,0,1 ,-1]

# MIN 은 승객을 찾는데 최소거리 , ans 는 DFS 함수에서 두 좌표 최단거리 
MIN = 1000
ans = 1000

# DFS 로 두 좌표간에 최단거리 찾기
def searchDFS(pos1 , pos2 , Marking_Arr , distance):
    
    np = pos1
    gp = pos2
    global ans
    Marking_Arr[np[0]][np[1]] = 1
    
    # 종료 조건
    if np[0] == gp[0] and np[1] == gp[1]:
        if distance < ans:
            ans = distance
            return
        
    for i in range(4):
        new_x = np[0] + action_dx[i]
        new_y = np[1] + action_dy[i]
    
        if new_x >= 0 and new_x < N and \
            new_y >= 0 and new_y < N and \
            Arr[new_x][new_y] != 1 and \
            Marking_Arr[new_x][new_y] == 0 :
           
            new_Marking_Arr = copy.deepcopy(Marking_Arr)
            searchDFS([new_x,new_y] , pos2 , new_Marking_Arr , distance + 1 )
    

# DFS 를 이용해서 최단거리가 제일 가까운 승객 찾기
def search_client( pos ):
    
    MIN = 100000
    global ans
    min_idx = -1
    i = 0
    
    for client in clients:
        ans = 100000
        Marking_Arr = [ [ 0 for _ in range(N)] for _ in range(N) ]
        searchDFS(pos , [ client[0] , client[1] ] , Marking_Arr , 0)
        
        if MIN > ans:
            MIN = ans
            min_idx = i  
            
        i = i + 1
    
    #print("최단 승객 : {} , 최단 거리 : {}  ".format(min_idx, MIN) )
    
    return min_idx , MIN

# 연료 충전?

end_flag = 0


# 메인 시뮬레이션 함수 : 연료/현재위치 업데이트하면서 진행
for i in range(M):
    
    print(now_pos)
    
    # 현재위치 - 고객 까지 거리
    client , distance_1 = search_client(now_pos)
    
    # 고객 이동 거리
    ans = 10000
    Marking_Arr = [ [ 0 for _ in range(N)] for _ in range(N) ]
    #print([clients[client][0] ,clients[client][1]] , [clients[client][2] ,clients[client][3]] )
    searchDFS( [clients[client][0] ,clients[client][1]] , [clients[client][2] ,clients[client][3]] , Marking_Arr , 0)
    distance_2 = ans
    
    if F < distance_1 + distance_2:
        end_flag = 1
        break
    else:
        now_pos[0] = clients[client][2]
        now_pos[1] = clients[client][3]
        F -= (distance_1 + distance_2)
        F += distance_2*2
        
    clients.pop(client)
        
if end_flag == 1:
    print(-1)
else:
    print(F)
```

```python


# 19237 . 어른 상어

import numpy as np

def minus(i):
    return int(i) - 1

# 테스트용 입력
N , M , K  =  map(int , input().split())
Input = list(map(int , input().split()))
Arr = np.array(Input).reshape(N,N).tolist()
directions = list(map(minus , input().split()))

Input = list(map(int , input().split()))
prior = (np.array(Input) - 1).reshape(4*M,4).tolist()
end_flag = M

# # 입력
# N , M , K  =  map(int , input().split())
# Arr = [list(map(int , input().split())) for _ in range(N)]
# directions = list(map(minus , input().split()))
# prior = [list(map(int , input().split())) for _ in range(4*M)]
# end_flag = M

# Arr 은 처음에만 씀

# 데이터 - 상어 위치 배열 , 상어 냄새 배열 
Occupy = [ [ [] for _ in range(N)] for _ in range(N)]
pos = []
Smell_Arr = [ [ [] for _ in range(N)] for _ in range(N)]

for x in range(N):
    for y in range(N):
        if Arr[x][y] != 0:
            shark = Arr[x][y] - 1
            pos.append( [shark , x , y , directions[shark] ])
            Occupy[x][y].append(shark)
            Smell_Arr[x][y].append( [shark, 4])
            
# 상어 위치정보는 - 상어 순서에 따라 정렬하기
pos = sorted(pos, key=lambda x: x[0], reverse=False)


# 상어 이동 우선순위

# 상어가 움직이는 규칙
action_dx = [-1, 1 ,0,0]
action_dy = [0,0,-1,1]

# 종료 조건

# 메인 시뮬레이션 함수
for time in range(1000):
    
    
    # 액션 1 : 상어 이동
    for idx in range(M):
        
        # 상어 위치 정보 업데이트
        shark = pos[idx][0]
        
        # 죽은 상어는 움직일수 없다
        if shark == -1 :
            continue
            
        x = pos[idx][1]
        y = pos[idx][2]
        direction =  pos[idx][3]
        move_flag = 1
        
        # 상어 근처 냄새가 없는 칸이 있을 때
        for i in prior[ shark*4 + direction ] :
            new_x = x + action_dx[i]
            new_y = y + action_dy[i]
            
            if new_x < 0 or new_x >= N or new_y < 0 and new_y >= N:
                continue
            
            if len(Smell_Arr[new_x][new_y]) == 0:
                pos[idx][1] = new_x
                pos[idx][2] = new_y
                pos[idx][3] = i
                Occupy[x][y] = []
                Occupy[new_x][new_y].append(shark)
                move_flag = 0
                break
                
        # 상어 근처 자기 냄새 기준으로 움직일 때
        if move_flag == 1:    
            for i in prior[ shark*4 + direction ] :
                new_x = x + action_dx[i]
                new_y = y + action_dy[i]
                if new_x < 0 or new_x >= N or new_y < 0 and new_y >= N:
                    continue
                    
                if Smell_Arr[new_x][new_y][0][0] == shark :
                    pos[idx][1] = new_x
                    pos[idx][2] = new_y
                    pos[idx][3] = i
                    Occupy[x][y] = []
                    Occupy[new_x][new_y].append(shark)
                    move_flag = 0
                    break
        
    # 액션 2 : 상어간의 혈투
    for x in range(N):
        for y in range(N):
            
            if len(Occupy[x][y]) > 1:
                
                # 누가 살아있을까
                alive_index = Occupy[x][y].index(max(Occupy[x][y]))
                alive = Occupy[x][y].pop(alive_index)
                
                # 죽일놈은 죽이고 업데이트
                delete = Occupy[x][y]
                for shark in Occupy[x][y]:
                    pos[shark][0] = -1
                    end_flag -= 1
                    
                Occupy[x][y] = [alive]
            
            # 상어 냄새 빠짐
            if len(Smell_Arr[x][y]) != 0:
                Smell_Arr[x][y][1] = Smell_Arr[x][y][1] - 1
                
                if Smell_Arr[x][y][1] == 0:
                    Smell_Arr[x][y] = []
                
    
    # 액션 3 : 상어 냄새 업데이트
    for idx in range(M):
        
        shark = pos[idx][0]
        # 죽은 상어는 움직일수 없다
        if shark == -1 :
            continue
            
        x = pos[idx][1]
        y = pos[idx][2]
        
        Smell_Arr[x][y].append([shark , 4])
        
    # 종료 조건
    if end_flag == 1:
        print(time + 1)
        
    # 턴 진행 : 상어 냄새 업데이트
    

if end_flag != 1:
    print(-1)





```
